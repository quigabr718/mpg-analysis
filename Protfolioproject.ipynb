```python
import seaborn as sns
sns.set_theme()
```

```python
mpg = sns.load_dataset("mpg")
```

```python
mpg
```

```python
sns.relplot(
    data = mpg,
    y = "mpg", x = "horsepower").set(title='Relationship between Hosepower and fuel economy (mpg)')
```

### When looking at horesepower and how it relates to fuel economy we see a negative correlation between the two factors. This means the higher the horesepower the miles per galon that the car has is lower. Car with a lower horsepower tend to have a higher fuel economy getting more miles to the gallon.

```python
carorigin = mpg.groupby(["origin"])["mpg"].mean()
```

```python
sns.set_theme()

sns.lineplot (
    data = mpg,
    x = "model_year", y= "mpg", errorbar= None).set(title='Relationship between Year of Model and Fuel Economy (mpg)')
```

### Over time fuel economy has gotten better. Model year and fuel economy have a positive correlation. This would make sense as new technology has improved fuel economy as a whole. There have been some spikes and dips along the way but on average, if we look at a cars model year and how much miles per gallon the car can do, the higher the model year, the higher the miles per gallon.

### The origin of the car does relate to the fuel economy. On average, cars made in Japan have a higher mpg, meaning it has a better fuel economy then Europe and the USA. Cars made in the US have a lower fuel economy then both other regions. 

```python
sns.set_theme()

sns.catplot (
    data = mpg,
    x = "origin", y= "mpg", kind= "bar", errorbar= None).set(title='Country of Orgin vs level of Fuel Economy (mpg)')
```

```python
sns.set_theme()

sns.relplot (
    data = mpg,
    x = "weight", y = "mpg").set(title='Relationship between Car Weight and Fuel Economy (mpg)')
```

### Similar to horsepower, the weight of the car also has a negative correlation with fuel economy. The heavier cars tend to get less miles to the gallon then cars that are lighter. 

```python
sns.set_theme()

sns.catplot (
    data = mpg,
    x = "cylinders", y= "mpg", kind= "bar", errorbar= None).set(title='Number of Cylinders vs level of Fuel Economy (mpg)')
```
